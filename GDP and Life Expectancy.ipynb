# Examining the Global Relationship between GDP per Capita and Life Expectancy in the year 2007 

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('countries.csv')
data
```

```python
data_2007 = data[data['year'] == 2007]
data_2007.head()
```

 # Relationship between GDP per capita and life expectancy
 

```python
plt.scatter(data_2007.gdpPerCapita,data_2007.lifeExpectancy,10)
plt.xlabel('GDP per Capita')
plt.ylabel('Life Expectancy')
plt.title ('Scatter Plot for GDP per Capita and Life Expectancy')
plt.show()
```

```python
data_2007.gdpPerCapita.corr(data_2007.lifeExpectancy)
```

```python
np.log10([10,100,1000])
```

```python
plt.scatter(np.log10(data_2007.gdpPerCapita),data_2007.lifeExpectancy)
plt.show()
```

```python
np.log10(data_2007.gdpPerCapita).corr(data_2007.lifeExpectancy)
```

```python
years_sorted = sorted(set(data.year))
years_sorted
```

## Run analysis for each available year

### Visualization of how each year performed

```python
for given_years in years_sorted:
    data_year = data[data.year == given_years]
    plt.scatter(data_year.gdpPerCapita,data_year.lifeExpectancy,5)
    plt.title (given_years)
    #limit the outliers
    plt.xlim(0,60000)
    plt.ylim(25,85)
    plt.xlabel('GDP per Capita')
    plt.ylabel('Life Expectancy')
  
   # plt.savefig(str(given_years),dpi = 500) #dpi =dots per inch
    #plt.clf()
    plt.show()

    
    
```

```python
for given_years in years_sorted:
    data_year = data[data.year == given_years]
    plt.scatter(np.log10(data_year.gdpPerCapita),data_year.lifeExpectancy,5)
    plt.title (given_years)
    #limit the outliers
    
    plt.xlabel('GDP per Capita')
    plt.ylabel('Life Expectancy')

   # plt.savefig('Log_' + str(given_years)) #dpi =dots per inch
    #plt.clf()
    plt.show()
    

    
```
